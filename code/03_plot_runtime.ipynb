```python
fig, ax = plt.subplots()

# === Color & Style Definitions ===
# Colors distinguish the specific implementation/variant
c_vanilla = "tab:blue"
c_parallel = "tab:red"
c_parallel_lb = "tab:green"

# === Plotting PSED (Solid Lines) ===
ax.plot(x, psed_vanilla_s, color=c_vanilla, marker="o", linestyle="-", label=r"PSED_vanilla")
ax.plot(x, psed_parallel_bsf_s, color=c_parallel, marker="^", linestyle="-", label=r"PSED_parallel_bsf")
ax.plot(x, psed_parallel_bsf_lb_s, color=c_parallel_lb, marker="s", linestyle="-", label=r"PSED_parallel_bsf_lb")

# === Plotting PSDTW (Dashed Lines) ===
ax.plot(x, psdtw_vanilla_s, color=c_vanilla, marker="o", linestyle="--", label=r"PSDTW_vanilla")
ax.plot(x, psdtw_parallel_bsf_s, color=c_parallel, marker="^", linestyle="--", label=r"PSDTW_parallel_bsf")
ax.plot(x, psdtw_parallel_bsf_lb_s, color=c_parallel_lb, marker="s", linestyle="--", label=r"PSDTW_parallel_bsf_lb")

ax.set_xlabel("Scaling Factor $l$")
ax.set_ylabel("Runtime (s)")

# ax.set_yscale("log")

ax.set_xticks([1.25, 1.50, 1.75, 2.0])

ax.grid(True, which="both", ls="--", lw=0.5, alpha=0.6)

# === Legend ===
# Placed outside to prevent covering data if the plot is crowded
ax.legend(loc="upper left")

plt.tight_layout()
plt.savefig("../figures/runtime_P2_second.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
