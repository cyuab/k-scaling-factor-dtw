```python
import matplotlib.pyplot as plt
```

```python
x = [1.25, 1.50, 1.75, 2.0]

plt.rcParams.update(
    {
        "font.size": 14,
        "axes.labelsize": 16,
        "axes.titlesize": 16,
        "legend.fontsize": 14,
        "xtick.labelsize": 14,
        "ytick.labelsize": 14,
        "figure.figsize": (7, 5),
        "lines.linewidth": 2.2,
    }
)
```

# P = 2

```python
psed_s = [21.49611496925354, 193.41970109939575, 618.812872171402, 1130.15372300148]
psed_lb_s = [
    36.78118634223938,
    315.9498248100281,
    1087.714034795761,
    1992.3721692562103,
]
psdtw_s = [615.3426630496979, 8077.524085283279, 30334.72380042076, 34813.23771739006]
psdtw_lb_s = [
    384.10391545295715,
    4735.816817998886,
    22524.8298227787,
    33017.17446565628,
]

psed_iter = [30741150, 274334823, 914233016, 1565817691]
psed_lb_iter = [29333287, 259045023, 869608741, 1506600699]
psdtw_iter = [32540659.0, 274994784.0, 880764560.0, 1513184734.0]
psdtw_lb_iter = [20321402.0, 208339178.0, 761803880.0, 1387359653.0]
```

```python
fig, ax = plt.subplots()

ax.plot(x, psed_s, marker="o", linestyle="-", label=r"PSED")
ax.plot(x, psed_lb_s, marker="s", linestyle="--", label=r"PSED ($LB_{\mathrm{Shen}}$)")
ax.plot(x, psdtw_s, marker="^", linestyle="-", label=r"PSDTW")
ax.plot(
    x, psdtw_lb_s, marker="D", linestyle="--", label=r"PSDTW ($LB_{\mathrm{Shen}}$)"
)

ax.set_xlabel("Scaling Factor $l$")
ax.set_ylabel("Runtime (s)")

# ax.set_yscale("log")

ax.set_xticks([1.25, 1.50, 1.75, 2.0])

ax.grid(True, which="both", ls="--", lw=0.5, alpha=0.6)

# === Legend ===
ax.legend(loc="upper left")

plt.tight_layout()
plt.savefig("../figures/runtime_P2_second.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
